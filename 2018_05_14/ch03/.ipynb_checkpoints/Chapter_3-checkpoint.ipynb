## ステップ関数のグラフ

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def function_step(x):
    return np.array(x > 0, dtype=np.int)

x = np.arange(-5.0, 5.0, 0.1)
y = function_step(x)
plt.plot(x, y)
plt.ylim(-0.1, 1.1) #y軸の範囲を指定する
plt.show()
```

## シグモイド関数の実装

```python
def function_sigmoid(x):
    return 1/(1 + np.exp(-x))

x = np.arange(-5.0, 5.0, 0.1)
y = function_sigmoid(x)
plt.plot(x, y)
plt.ylim(-0.1, 1.1) #y軸の範囲を指定する
plt.show()
```

```python
def function_relu(x):
    return np.maximum(0,x)
```

## 多次元配列の計算

## 3層ニューラルネットワークの実装

### 入力層から第1層目への信号の伝達

```python
X = np.array([1.0, 0.5])
W1 = np.array([[0.1, 0.3, 0.5], [0.2, 0.4, 0.6]])
B1 = np.array([0.1, 0.2, 0.3])

A1 = np.dot(X, W1) + B1 #入力層から第1層へのinput = 隠れ層での重み付き和
Z1 = function_sigmoid(A1) #シグモイド関数の活性化関数で信号変換

print(A1)
print(Z1)
```

### 第１層から第2層への信号伝達

```python
W2 = np.array([[0.1, 0.4], [0.2, 0.5], [0.3, 0.6]]) #重みが変わる
B2 = np.array([0.1, 0.2]) #バイアスも変わる

A2 = np.dot(Z1, W2) + B2 #前の層で受け取った信号を入力とする
Z2 = function_sigmoid(A2)

print(A2)
print(Z2)
```

### 第2層から出力層への信号伝達

```python
def function_identify(x):
    return x

W3 = np.array([[0.1, 0.3], [0.2, 0.4]])
B3 = np.array([0.1, 0.2]) 

A3 = np.dot(Z2, W3) + B3
y = function_identify(A3)

print(y)
```

## 出力層の設計

### ソフトマックス関数の実装

```python
def function_softmax_origin(x):
    exp_x = np.exp(x)
    sum_exp_x = np.sum(exp_x)
    y = exp_x / sum_exp_x
    
    return y
```

#### 指数関数を使うので値が大きくなるとオーバーフローが発生し、結果は不安定になる

### オーバーフロー対策をしたソフトマックス関数の実装

```python
def function_softmax(x):
    exp_x = np.exp(x)
    sum_exp_x = np.sum(exp_x)
    y = exp_x / sum_exp_x
    return y
```
